```python
import sys
from qgis.core import *
from qgis.PyQt.QtGui import *
from qgis.PyQt.QtCore import *
QgsApplication.setPrefixPath(r'C:\\OSGeo4W64\\apps\\qgis', True)
qgs = QgsApplication([], True)
qgs.initQgis()
sys.path.append(r'C:\OSGeo4W64\apps\qgis\python\plugins')
project = QgsProject.instance()
```

```python
csv_path = r"file:///C:\\Users\\ASUS\\Desktop\\Trabalho_SIG\\US\\us-states-todasasdatas.csv?delimiter=,'"  
csv_layer = QgsVectorLayer(csv_path, 'Data', 'delimitedtext')

if not csv_layer.isValid():
    print('Layer failed to load!')
else:
    project.addMapLayer(csv_layer)
    print("Sucess")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.rcParams.update(plt.rcParamsDefault)
plt.style.use('ggplot')
values = []
labels = []
datas = []
lowerTotal = 0
i = 0


for feature in csv_layer.getFeatures():
    if feature['state']=="Connecticut":
        if feature['cases']>0:
            values.append(feature['cases'])
            labels.append(feature['date'])
            

fig = plt.figure(figsize=[30,10]);

cov_date = pd.Series(values, index=labels)
plt.xlabel('Data (Dia-Ms-Ano)', fontsize=50)
plt.tick_params(axis='x', rotation=70)
plt.ylabel('Nmero de casos', fontsize=50)
plt.title('Nmero de casos em Connecticut ao longo do tempo', fontsize=50)
plt.plot(cov_date, '.-')
plt.savefig('Figuras/casoscon.jpg', bbox_inches='tight')
plt.show()
```
